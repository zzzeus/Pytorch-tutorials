```python
# show Sigmoid
m = nn.Sigmoid()
input = torch.randn(1000)
output = m(input)
plt.plot(input.numpy(),output.numpy(),'.')
plt.show()
```

```python
# Show ReLU
m = nn.ReLU()
input = torch.randn(1000)
output = m(input)
plt.plot(input.numpy(),output.numpy(),'.')
plt.show()
```

```python
# Show ELU
m = nn.ELU()
input = torch.randn(1000)
output = m(input)
plt.plot(input.numpy(),output.numpy(),'.')
plt.show()
```

```python
# Show LeakyReLU
m = nn.LeakyReLU()
input = torch.randn(1000)
output = m(input)
plt.plot(input.numpy(),output.numpy(),'.')
plt.show()
```

```python
# Show Tanh
m = nn.Tanh()
input = torch.randn(1000)
output = m(input)
plt.plot(input.numpy(),output.numpy(),'.')
plt.show()
```
